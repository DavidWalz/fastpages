```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
def plot():
    fig, ax = plt.subplots(1)
    ax.plot(x, y)

plot()
```

```python
def plot():
    fig, ax = plt.subplots(1)
    ax.plot(x, y)
    return fig

plot()
```
